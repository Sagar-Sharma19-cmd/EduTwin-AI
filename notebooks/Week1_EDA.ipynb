Label Encoding (Categorical → Numbers)
ML models only understand numbers — not text like "yes", "no", "GP", "F". So we need to convert all text columns into numbers. This is called Label Encoding.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Load clean dataset
df = pd.read_csv('../data/student_clean.csv')

print("Dataset shape:", df.shape)
print("\nColumn types:")
print(df.dtypes)
```

```python
# Find all text (object) columns
text_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Text columns that need encoding:")
print(text_cols)
print(f"\nTotal: {len(text_cols)} columns to encode")
```

```python
# Apply Label Encoding to all text columns
le = LabelEncoder()
df_encoded = df.copy()

for col in text_cols:
    df_encoded[col] = le.fit_transform(df[col])
    print(f"✅ {col:15} → encoded  | Unique values: {df[col].unique()}")

print("\n✅ All text columns encoded!")
print("New dtypes:")
print(df_encoded.dtypes.value_counts())
```

Feature Scaling
Some features have values 0-4 (like studytime) and others have 0-93 (like absences). This imbalance confuses ML models. We fix it with StandardScaler — makes everything on the same scale!


```python
# Columns to scale (all numeric except target variables)
target_cols = ['G3', 'pass_fail']
scale_cols = [col for col in df_encoded.select_dtypes(
              include=[np.number]).columns 
              if col not in target_cols]

print("Columns to scale:")
print(scale_cols)
```

```python
# Apply StandardScaler
scaler = StandardScaler()
df_scaled = df_encoded.copy()
df_scaled[scale_cols] = scaler.fit_transform(df_encoded[scale_cols])

print("\nBefore scaling (absences):")
print(f"  Min: {df_encoded['absences'].min()}, Max: {df_encoded['absences'].max()}")

print("\nAfter scaling (absences):")
print(f"  Min: {df_scaled['absences'].min():.2f}, Max: {df_scaled['absences'].max():.2f}")
print("\n✅ All features scaled to same range!")
```

```python
# Visualise before vs after scaling
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df_encoded['absences'], bins=20, 
             color='salmon', edgecolor='black')
axes[0].set_title('Absences — BEFORE Scaling')
axes[0].set_xlabel('Value')

axes[1].hist(df_scaled['absences'], bins=20, 
             color='steelblue', edgecolor='black')
axes[1].set_title('Absences — AFTER Scaling')
axes[1].set_xlabel('Scaled Value')

plt.tight_layout()
plt.savefig('../notebooks/chart6_scaling.png')
plt.show()
print("✅ Scaling chart saved!")
```

Feature Selection
Not all 33 columns are useful for predicting grades. We pick only the most important ones using correlation. This is called Feature Selection!

```python
# Find correlation of every feature with G3
correlation = df_encoded.corr()['G3'].abs().sort_values(ascending=False)

print("Feature correlation with Final Grade (G3):")
print(correlation.round(3))
```

```python
# Visualise feature importance
plt.figure(figsize=(10, 8))
correlation.drop(['G3', 'pass_fail']).plot(
    kind='barh',
    color=['steelblue' if v > 0.2 else 'lightgray' 
           for v in correlation.drop(['G3', 'pass_fail'])],
    edgecolor='black'
)
plt.title('Feature Correlation with Final Grade (G3)')
plt.xlabel('Correlation Score')
plt.axvline(x=0.2, color='red', linestyle='--', 
            label='Threshold (0.2)')
plt.legend()
plt.tight_layout()
plt.savefig('../notebooks/chart7_feature_importance.png')
plt.show()
print("✅ Feature importance chart saved!")
```

```python
# Select top features (correlation > 0.1 with G3)
top_features = correlation[correlation > 0.1].index.tolist()
top_features = [f for f in top_features if f != 'G3']

print(f"\n✅ Selected {len(top_features)} important features:")
print(top_features)
```

Train/Test Split
We split data into 80% training (model learns from this) and 20% testing (we test how good the model is). This is super important to avoid overfitting!


```python
# Prepare X (features) and y (target)
X = df_scaled[top_features]
y_regression     = df_scaled['G3']        # for regression (Week 3)
y_classification = df_scaled['pass_fail'] # for classification (Week 4)

print("X shape (features):", X.shape)
print("y regression shape:", y_regression.shape)
print("y classification shape:", y_classification.shape)
```

```python
# Split into train and test — 80/20
X_train, X_test, y_train_reg, y_test_reg = train_test_split(
    X, y_regression, 
    test_size=0.2, 
    random_state=42
)

X_train, X_test, y_train_clf, y_test_clf = train_test_split(
    X, y_classification,
    test_size=0.2,
    random_state=42
)

print("✅ Train/Test Split Done!")
print(f"Training samples:  {X_train.shape[0]} ({X_train.shape[0]/len(X)*100:.0f}%)")
print(f"Testing samples:   {X_test.shape[0]}  ({X_test.shape[0]/len(X)*100:.0f}%)")
```

```python
# Visualise the split
plt.figure(figsize=(6, 4))
plt.pie([len(X_train), len(X_test)],
        labels=['Training (80%)', 'Testing (20%)'],
        colors=['steelblue', 'salmon'],
        autopct='%1.0f%%',
        startangle=90)
plt.title('Train / Test Split')
plt.savefig('../notebooks/chart8_train_test_split.png')
plt.show()
print("✅ Split chart saved!")
```

```python
# Save all preprocessed data
df_encoded.to_csv('../data/student_encoded.csv', index=False)
df_scaled.to_csv('../data/student_scaled.csv', index=False)

# Save train/test sets
X_train.to_csv('../data/X_train.csv', index=False)
X_test.to_csv('../data/X_test.csv', index=False)
y_train_reg.to_csv('../data/y_train_reg.csv', index=False)
y_test_reg.to_csv('../data/y_test_reg.csv', index=False)
y_train_clf.to_csv('../data/y_train_clf.csv', index=False)
y_test_clf.to_csv('../data/y_test_clf.csv', index=False)

print("✅ All preprocessed files saved!")
print("\nFiles created:")
import os
for f in os.listdir('../data/'):
    print(f"  📄 {f}")
```
